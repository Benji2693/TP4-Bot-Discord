TP 4 - Bot Discord

1) Qu’est-ce qu’un bot ?

Un bot n'est pas humain et effectue des taches qu'une personne pourrait faire de facon programmé.
C'est donc une application programmée pour effectuer des taches automatiquement.

2) Qu'est ce que le logging. À quoi servent les logs?

Le logging est la journalisation. C'est donc la gestion des logs qui sont des journaux d'evenement qui est représenté par un fichier ou tout autre élement pouvant contenir l'historique des evenements. 
Les logs servent à suivre les evenements qui se produisent lors de l'exécution de certains logiciels. L'ajout d'appels de journalisation au code permet d'indiquer que certains événements se sont produits. Un événement est décrit par un message descriptif qui peut éventuellement contenir des données variables.



Cet ecemple de logging permet de voir si nous somme bien rentré dans la fonction add. Dans la variable level on vas afficher plus ou moins de détails sur les élements et envoyé dans un fichier appart comme on peut voir sur la deuxième ligne.
